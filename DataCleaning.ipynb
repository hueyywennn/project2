```python
# Drop rows where all values are NaN
df.dropna(how='all', inplace=True)

df
```

```python
# convert the date column into a datetime object
df['Date Flown'] = pd.to_datetime(df['Date Flown'])

# Split the "Date Flown" column into "Month", and "Year" columns
if 'Date Flown' in df.columns:
    # extract the day, month, and year components
    df['Month Flown'] = df['Date Flown'].dt.month
    df['Month Flown'] = df['Month Flown'].fillna(0).astype(int)
    df['Year Flown'] = df['Date Flown'].dt.year
    df['Year Flown'] = df['Year Flown'].fillna(0).astype(int)
else:
    df['Month Flown'] = "0"
    df['Year Flown'] = "0"

df
```

```python
# Replace all occurrences of "N/A" with NaN
#df.replace("N/A", "Unknown", inplace=True)
df.fillna("0", inplace=True)

df
```

```python
import re

# Regex pattern to remove "Trip Verified" or "Not Verified" along with any leading "|"
pattern = r"\s?(✅?\s?Trip Verified|Not Verified)\s?\|?"

# Apply the cleaning function to the "review" column
df['Review'] = df['Review'].apply(lambda x: re.sub(pattern, "", str(x)))

df.head()
```

```python
# Split the "Route" column into "Departures", "Arrivals", and "Layover" columns
if 'Route' in df.columns:
    new = df['Route'].str.extract(r'^(.*?) to (.*?)(?: via (.*))?$')
    df["Departures"] = new[0]
    df["Arrivals"] = new[1]
    df["Layover"] = new[2]
else:
    df["Departures"] = "Empty"
    df["Arrivals"] = "Empty"
    df["Layover"] = "Empty"

# Ensure any NaN values in these columns are replaced with "Empty"
df[["Departures", "Arrivals", "Layover"]] = df[["Departures", "Arrivals", "Layover"]].fillna("Empty")

df
```

```python
df = df.drop(columns=['Date', 'Author'])
```

```python
df.to_csv("cleaned_data_part1.csv", index=False)

print("Data cleaned and saved to 'cleaned_data_part1.csv'")
```
